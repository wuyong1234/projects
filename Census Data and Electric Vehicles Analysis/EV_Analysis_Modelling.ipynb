```python
import seaborn as sb
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
warnings.filterwarnings("ignore")  # ignore runtime warning

# outcome 
outcomes = df.pop('outcome')

# filter out unwanted columns
features = df
```

```python
# Split data into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, outcomes, test_size=0.3, random_state=1000, stratify = outcomes)

# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from lazypredict.Supervised import LazyClassifier  # import regression 

clf = LazyClassifier(verbose = 0, predictions = True) 
# verbose = 0 : no print out 
# predictions = True : print the prediction result for test set

models, predictions = clf.fit(X_train, X_test, y_train, y_test)
models
```

### Algorithms chosen

```python
model_chosen = ['LogisticRegression', 'RandomForestClassifier', 'SVC']
```

### Algorithms Verification

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, fbeta_score
from sklearn.metrics import classification_report

def score_gen(report, model_name):    
    lines = report.split('\n')
    score = {}
    score['model'] = model_name 
    for i in range(len(lines[5:-1])):
    #     print(lines[5:-1][i].split('      '))
        if i == 0:
            score['accuracy'] = float(lines[5:-1][i].split('      ')[4].strip())*100
        elif i == 2:
            score['precision'] = float(lines[5:-1][i].split('      ')[1].strip())*100
            score['recall'] = float(lines[5:-1][i].split('      ')[2].strip())*100
            score['f1-score'] = float(lines[5:-1][i].split('      ')[3].strip())*100
            score['support'] = float(lines[5:-1][i].split('      ')[4].strip())
    return pd.DataFrame([score])
 
report_1 = classification_report(y_test, predictions[model_chosen[0]])
report_2 = classification_report(y_test, predictions[model_chosen[1]])
report_3 = classification_report(y_test, predictions[model_chosen[2]])

score_lr = score_gen(report_1, model_chosen[0])
score_rf = score_gen(report_2, model_chosen[1])
score_svc = score_gen(report_3, model_chosen[2])

frames = [score_lr, score_rf, score_svc]
score = pd.concat(frames).reset_index(drop = True)
score
```

```python
import numpy as np 
import matplotlib.pyplot as plt 
  
# for SVM, the highest values from the kernel, poly
X   = score.model.tolist()

#Collect this information from the classification reports and store them in lists
accuracy  = score.accuracy.tolist()
precision = score.precision.tolist()
recall    = score.recall.tolist()
f1        = score['f1-score'].tolist()

#here we use np.arange to get the number of items to place on the x-axis 
X_axis = np.arange(len(X))

#the spacing between each bar each bar
spacing= 0.15

#setting the size of the graph
plt.rcParams["figure.figsize"] = (16,7)
  
#below we are plotting 4 bars for each label on the x-axis with the following syntax
#plt.bar(star_pos_for_each_bar, list_of_values, width_of each_bar, label_name_for each bar)
plt.bar(X_axis, accuracy, 0.25, label = 'Accuracy')
plt.bar(X_axis + spacing, precision, 0.25, label = 'Precision')
plt.bar(X_axis + spacing *2, recall, 0.25, label = 'Recall')
plt.bar(X_axis + spacing *3, f1, 0.25, label = 'F1')

#setting the range of values for the X axis
plt.xticks(X_axis, X)
plt.grid(axis = 'y')
plt.xlabel("Model")
plt.ylabel("Weighted Avg")
plt.title("Model vs Weighted Avg Of Various Metrics")
plt.legend(loc='upper right')

plt.show()
```

## Accuracy check per algorithms

```python
from sklearn.metrics import accuracy_score, accuracy_score, recall_score, f1_score, fbeta_score
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from skopt import BayesSearchCV
from lazypredict.Supervised import LazyClassifier  # import regression if needed 
from sklearn.linear_model import LogisticRegression
import tensorflow as tf
from tensorflow import keras


def pred_logi(X_train, X_test, y_train, y_test, pred_features):
    pipe = Pipeline(steps = [('classifier', LogisticRegression())])

    pipe.fit(X_train, y_train)
    y_pred = pipe.predict(X_test)
    
    accuracy = accuracy_score(y_pred, y_test)
    prediction = int(pipe.predict(pred_features))
    print(prediction)
    
    return [accuracy, prediction]


def pred_svc(X_train, X_test, y_train, y_test, pred_features):
    pipe = Pipeline(steps = [('classifier', SVC())])

    # https://scikit-optimize.github.io/stable/modules/generated/skopt.BayesSearchCV.html    
    # w/ tuning
    opt = BayesSearchCV(
        pipe, 
        {
            'classifier__C': (1e-1, 1e+1, 'log-uniform'), 
            'classifier__gamma': (1e-6, 1e+1, 'log-uniform'),
            'classifier__degree': (1, 2), # was (1, 8)
            'classifier__kernel': ['linear', 'poly', 'rbf'], 
        }, 
        n_iter = 3, 
        cv = 3
    )
    
    opt.fit(X_train, y_train)
    y_pred = opt.predict(X_test)
    
    accuracy = accuracy_score(y_pred, y_test)
    prediction = int(opt.predict(pred_features))  
    print(prediction)
    
    return [accuracy, prediction]


def pred_rf(X_train, X_test, y_train, y_test, pred_features):
    clf_rf = RandomForestClassifier()

    # Set up the hyperparameter search
    param_dist = {"max_depth": [3, None],
                  "n_estimators": list(range(1, 10)),
                  "max_features": list(range(1, X_test.shape[1]+1)),
                  "min_samples_split": list(range(2, 11)),
                  "min_samples_leaf": list(range(1, 11)),
                  "bootstrap": [True, False],
                  "criterion": ["entropy"]}

    # Run a randomized search over the hyperparameters
    random_search = RandomizedSearchCV(clf_rf, param_distributions=param_dist)

    # Fit the model on the training data
    random_search.fit(X_train, y_train)
    y_pred = random_search.best_estimator_.predict(X_test)

    accuracy = accuracy_score(y_pred, y_test)
    prediction = int(random_search.best_estimator_.predict(pred_features))
    print(prediction)
     
    return [accuracy, prediction]


def classifer_data_input(df):
    # features to predict
    df.drop(columns = ['ID'], inplace = True)
    pred_features = df[-1:].drop(columns = 'outcome')
    
    # features to train &test
    df = df[:-1]

    # outcome 
    outcomes = df.pop('outcome').values
    # filter out unwanted columns
    features = df.values
    
    # for non-tf
    X_train, X_test, y_train, y_test = train_test_split(features, outcomes, test_size=0.2, random_state= 1000, stratify = outcomes)

    # standarization
    sc = StandardScaler()
    X_train = sc.fit_transform(X_train)
    X_test = sc.transform(X_test)
    
    logi = pred_logi(X_train, X_test, y_train, y_test, pred_features)
    svc = pred_svc(X_train, X_test, y_train, y_test, pred_features)
    rf = pred_rf(X_train, X_test, y_train, y_test, pred_features)
    
    result = {}
    result['logi'] = [logi[1]]
    result['svc'] = [svc[1]]
    result['rf'] = [rf[1]]

    accuracy = {}
    accuracy['logi'] = [logi[0]]
    accuracy['svc'] = [svc[0]]
    accuracy['rf'] = [rf[0]]

    rank = pd.DataFrame(accuracy).T.rename(columns = {0 :'Accuracy'}).sort_values('Accuracy', ascending = False)
    return rank
```
