```python
# One hot encoding our gender column to convert it from categorical to numerical
pd.get_dummies(df['gender'], drop_first=True)
```

```python
# Converting the data in the gender column to 0s and 1s
df['gender'] = pd.get_dummies(df['gender'], drop_first = True)
```

```python
df.head()
```

```python
# Separting the Features from the labels, Separating X from y 
X = df[['long_hair','forehead_width_cm','forehead_height_cm','nose_wide','nose_long','lips_thin','distance_nose_to_lip_long']]
X
```

```python
# Extracting the labels
y = df[['gender']]
y
```

```python
# Split your data into training and testing sets

from sklearn.model_selection import train_test_split

# Splitting the dataset into the training and test sets - 70% of the data is used for training and 30% is used for testing

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state = 10)
```

```python
# Creating a Linear Regression Model

from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()

# Training the Linear Regression model using the training data
lr.fit(X_train, y_train)
```

```python
# Making predictions on the Test dataset
y_pred = lr.predict(X_test)
```

```python
# Generating the confusion matrix using sklearn's confusion matrix method
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, y_pred)
```

```python
#Generating the confusion matrix as a dataframe. 
pd.DataFrame( confusion_matrix(y_test, y_pred), columns = ['Actual Positive (male)', 'Acutal Negative (female)'], index = [ 'Predicted Positive (male)', 'Predicted Negative (female)']          )
```

```python
# Generate Classification Report
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

```python
# Converting the prediction to a dataframe
y_pred_df = pd.DataFrame(y_pred, columns = ['Predicted Gender'])
y_pred_df
```

```python
# Creating a merged dataframe of test data and the predictions to show our predicted gender with the actual gender
df = pd.concat([X_test.reset_index(drop=True), y_test.reset_index(drop=True), y_pred_df.reset_index(drop=True)], axis=1)
df
```
