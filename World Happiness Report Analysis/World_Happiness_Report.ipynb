World Happiness Report 2019 

Here is the link to the dataset: https://www.kaggle.com/unsdsn/world-happiness?select=2019.csv


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(color_codes = True)
import warnings 
warnings.filterwarnings("ignore")
```

```python
report = pd.read_csv('2019.csv',)
```

```python
report.head()
```

```python
report.isnull().sum()
```

```python
report.plot('Score', 'GDP per capita')
```

Here we can see that there is a positive correlation between GDP per capita and happiness score.

```python
report.plot('Score', 'Social support')
```

Here we can see that there is a general positive correlation between happiness score and social support percieved by people.

```python
report.plot('Score', 'Healthy life expectancy')
```

Here we can see that there is a general positive correlation between happiness score and healthy life expectancy.

```python
report.plot('Score', 'Freedom to make life choices')
```

Here we can see that there is a general positive correlation between happiness score and freedom to make life choices.

```python
report.plot('Score', 'Generosity')
```

Here we can see that there might not be a relationship between perceived generosity and happiness level in people.

```python
report.plot('Score', 'Perceptions of corruption')
```

Here we can see that there might not be a relationship between perceived percentions of curruption and happiness level in people.

```python
boxplot_data= report.drop(['Overall rank','Country or region', 'Score'],axis=1)
boxplot_data.head()
```

```python
boxplot_data.plot.box(figsize=(10,5))
plt.xticks(rotation = 90)
```

Here we can see the outliers present in each category.

```python
sns.swarmplot(data=boxplot_data)
sns.violinplot(data=boxplot_data)
plt.xticks(rotation = 90)
```

Here we see the distribution of each category. 

```python
top5 = report.head(5) 
top5
```
